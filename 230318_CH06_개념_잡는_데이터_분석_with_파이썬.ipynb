## **(3) 데이터 가공 및 그룹핑**

### **1) 특정 칼럼 삭제**

```python
col = ['Aired Date', 'Duration', 'Synopsis', 'Cast', 'Tags']
newdrama = drama.drop(col, axis=1)
newdrama.columns
```

### **2) 칼럼 이름 변경**

```python
newdrama.columns = ['이름','개봉년도','요일','에피소드수','네트워크','등급','장르','순위','평점']
newdrama.columns
```

## **3) 칼럼 생성하기**

```python
newdrama['장르개수'] = newdrama['장르'].str.count(',')+1
newdrama.head()
```

```python
newdrama['네트워크개수'] = newdrama['네트워크'].str.count(',')+1
newdrama.head()
```

### **4) 컬럼 기준으로 그룹화하기**

```python
d = newdrama.groupby('개봉년도')['평점'].mean()
pd.DataFrame(d)
```

```python
d = newdrama.groupby('개봉년도')['이름'].count()
pd.DataFrame(d)
```

```python
d = newdrama.groupby(['개봉년도','평점'])['이름'].count()
pd.DataFrame(d).head(10)
```

## **(4) 데이터 시각화**

### **1) 선 그래프 그리기**

```python
import matplotlib.pyplot as plt
import numpy as np

data = newdrama.groupby('개봉년도')['평점'].mean()
plt.figure(figsize=(16,10))
plt.plot(data, color='r', marker='o')
plt.xticks(np.arange(2003, 2022))
plt.xlabel('개봉년도')
plt.ylabel('평점의 평균')
plt.grid()
plt.show()
```

### **2) 막대 그래프 그리기**

```python
import pandas as pd
import matplotlib.pyplot as plt

top5 = newdrama['개봉년도'].value_counts().head(5)

plt.figure(figsize=(16,10))
plt.bar(top5.index, top5, color='g')
plt.xlabel('개봉년도')
plt.ylabel('장르개수')
plt.title('개봉년도별 장르개수 TOP5')
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

year_Count = newdrama.groupby('개봉년도').size()
plt.figure(figsize=(16,10))
plt.plot(year_Count, color='r', marker='o')
plt.bar(year_Count.index, year_Count, color='g')
plt.xticks(np.arange(2003, 2022))
plt.xlabel('Year of release')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
from collections import Counter
import pandas as pd

newdrama['장르'] = newdrama['장르'].str.strip()
genre_list = []

for genres in newdrama['장르'].to_list():
  genres = genres.split(", ")
  for gen in genres:
    genre_list.append(gen)

# rename(columns=0:'Count') -> rename(columns={0:'Count'}), chatGPT가 찾아줌
genre_df = pd.DataFrame.from_dict(Counter(genre_list), orient='index').rename(columns={0:'Count'})
genre_df.sort_values('Count', ascending=False, inplace=True)
genre_df.head()
```

```python
plt.figure(figsize=(25, 10))
plt.plot(genre_df, color='r', marker='o')
plt.bar(genre_df.index, genre_df.Count)
plt.title('장르별 드라마 개수')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```
